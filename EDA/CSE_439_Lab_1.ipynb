```python
import glob
images = glob.glob('sample_data/*.jpg')
```

# loading the images

```python
N = len(images)
print(f"Il y a {N} images :")
for f in images:
    print(f)
```

# drawing plot area for each image

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1,N)
plt.show()
```

# displaying all images

```python
import skimage.io as io
import matplotlib.pyplot as plt

# Figure with 1 row, N columns, and a specific size
fig, axs = plt.subplots(1,N, figsize=(15,5))

# Counter for the axes
i = 0

# Loop over the images
for f in images:

    # Load the image
    img = io.imread(f)

    # Display the image in the axe i
    axs[i].imshow(img)

    # Add a title with the filename
    axs[i].set_title(f)

    # Increment the counter
    i = i + 1

# Finally, show the figure
plt.show()
```

```python
import skimage.color as color

fig, axs = plt.subplots(1,N, figsize=(15,5))
i=0
for f in images:
  img = io.imread(f)
  img = color.rgb2gray(img)
  axs[i].imshow(img)
  axs[i].set_title(f)
  i = i + 1
plt.show()

```

# converting images into greyscale

```python
fig, axs = plt.subplots(1,N, figsize=(15,5))
i=0
for f in images:
  img = io.imread(f)
  img = color.rgb2gray(img)
  axs[i].imshow(img, cmap="gray")
  axs[i].set_title(f)
  i = i + 1
plt.show()
```

# measuring maximum and minimum greyscale value for each image

```python
i=0
for f in images:
  img = io.imread(f)
  img = color.rgb2gray(img)
  axs[i].imshow(img, cmap="gray")
  axs[i].set_title(f)
  print(axs[i].get_title())
  i = i + 1
  print(f"Min : {img.min()}")
  print(f"Max : {img.max()}\n")
```

# testing with limiting greyscale value

```python
for f in images:
  img = io.imread(f)
  img = color.rgb2gray(img)
  print(img[0,:])
  print('\n')
```

```python
fig, axs = plt.subplots(1,N, figsize=(15,5))
i=0
for f in images:
  img = io.imread(f)
  img = color.rgb2gray(img)
  axs[i].imshow(img, vmin=0.0, vmax=0.5, cmap="gray");
  axs[i].set_title(f)
  i = i + 1
plt.show()
```

# checking brightness profile for an image

```python
test = io.imread(images[1])
g = color.rgb2gray(test)
cut = 400         # Cut location
profil = g[cut,:]  # Extract all the pixels of the row 'cut' in the image g
plt.figure(figsize=(10,5))           # New figure with specific display size

plt.subplot(1,2,1)                   # Plot on the left
plt.plot(profil,'b')                 # Brightness profile
plt.title("Brightness profile")      # Title

plt.subplot(1,2,2)                   # Image on the right
plt.imshow(g, cmap="gray")           # Image
plt.plot((0,1599), (cut,cut), 'b')    # Cut line in blue
plt.title("Mandrill")                # Title

plt.show()
```
